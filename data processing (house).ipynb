```python
df=df[df['price'] !='Price on Request']
```

```python
def treat_price(x):
    if type(x)==float:
        return x
    else:
        if x[1]=='Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df['price']=df['price'].str.split(' ').apply(treat_price)
```

```python
df['price_per_sqrt'] = df['price_per_sqrt'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
```
